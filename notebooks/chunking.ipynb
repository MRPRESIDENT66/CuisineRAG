# Chunking Component
This notebook implements and compares four chunking strategies for the CuisineRAG system
using the `langchain-text-splitters` library:
1. CharacterTextSplitter (Fixed-size chunking)
2. RecursiveCharacterTextSplitter (Smarter fixed-size chunking)
3. MarkdownHeaderTextSplitter (Section-based chunking)
4. JSON section-based chunking
5. Semantic chunking (Embedding-based chunking)

```python
from langchain_text_splitters import (
    CharacterTextSplitter,
    RecursiveCharacterTextSplitter,
    MarkdownHeaderTextSplitter
)
from semantic_chunking import semantic_chunk_documents
```

```python
import json
from pathlib import Path

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / 'data').exists() and (PROJECT_ROOT.parent / 'data').exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
```

## Load Corpus from Scraping Output

```python
corpus_path = PROJECT_ROOT / 'data/raw/south_asian_corpus.json'
if not corpus_path.exists():
    raise FileNotFoundError(
        f'Missing corpus file: {corpus_path}. Run scraping notebook first or verify the notebook working directory.'
    )

with corpus_path.open('r', encoding='utf-8') as f:
    corpus = json.load(f)

if not corpus:
    raise ValueError('Corpus is empty. Check scraping output before chunking.')


def build_markdown_from_sections(doc: dict) -> str:
    title = doc.get('title', 'Untitled')
    summary = doc.get('Summary', '')
    sections = doc.get('sections', []) or []

    parts = [f"# {title}"]
    if summary:
        parts.append(summary)

    for sec in sections:
        sec_title = (sec.get('section_title') or '').strip()
        sec_text = (sec.get('section_text') or '').strip()
        if not sec_text:
            continue
        if sec_title.lower() == 'summary':
            continue
        if ' / ' in sec_title:
            parent, child = sec_title.split(' / ', 1)
            parts.append(f"## {parent.strip()}")
            parts.append(f"### {child.strip()}")
            parts.append(sec_text)
        else:
            parts.append(f"## {sec_title}")
            parts.append(sec_text)

    return '\n\n'.join(parts)


docs_for_chunking = []
blog_section_docs = []
for doc in corpus:
    text = build_markdown_from_sections(doc)
    if text.strip():
        docs_for_chunking.append({
            'title': doc.get('title', ''),
            'url': doc.get('url', ''),
            'text': text,
        })

    summary = (doc.get('Summary') or '').strip()
    if summary:
        blog_section_docs.append({
            'title': doc.get('title', ''),
            'url': doc.get('url', ''),
            'section_title': 'Summary',
            'text': summary,
        })

    for sec in doc.get('sections', []) or []:
        sec_title = (sec.get('section_title') or '').strip()
        sec_text = (sec.get('section_text') or '').strip()
        if not sec_text or sec_title.lower() == 'summary':
            continue
        blog_section_docs.append({
            'title': doc.get('title', ''),
            'url': doc.get('url', ''),
            'section_title': sec_title,
            'text': sec_text,
        })

texts = [d['text'] for d in docs_for_chunking]
metadatas = [{'title': d['title'], 'url': d['url']} for d in docs_for_chunking]
semantic_texts = [d['text'] for d in blog_section_docs]
semantic_metadatas = [
    {
        'title': d['title'],
        'url': d['url'],
        'section_title': d['section_title'],
    }
    for d in blog_section_docs
]

print(f'Loaded {len(texts)} documents from {corpus_path}')
print(f"Total characters: {sum(len(t) for t in texts):,}")
print('\n--- First document preview (500 chars) ---')
print(texts[0][:500])
```

## 1. CharacterTextSplitter (Fixed-Size Chunking)
Splits text into fixed-size chunks based on character count with overlap.
Simple and reliable, works on any text format.

```python
char_splitter = CharacterTextSplitter(
    separator=" ",
    chunk_size=300,
    chunk_overlap=50,
    length_function=len,
)

char_chunks = char_splitter.create_documents(texts, metadatas=metadatas)

print(f"Number of chunks: {len(char_chunks)}\n")
for i, chunk in enumerate(char_chunks[:5]):
    print(f"[Chunk {i}] {len(chunk.page_content)} chars | title={chunk.metadata.get('title', '')}")
    print(chunk.page_content[:300])
    print("---")
```

## 2. RecursiveCharacterTextSplitter (Smarter Fixed-Size Chunking)
Tries to split on natural boundaries first (paragraphs → sentences → words)
before falling back to character splitting. Produces more meaningful chunks
than plain CharacterTextSplitter.

```python
recursive_splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=50,
    length_function=len,
    separators=["\n\n", "\n", ".", " ", ""],
)

recursive_chunks = recursive_splitter.create_documents(texts, metadatas=metadatas)

print(f"Number of chunks: {len(recursive_chunks)}\n")
for i, chunk in enumerate(recursive_chunks[:5]):
    print(f"[Chunk {i}] {len(chunk.page_content)} chars | title={chunk.metadata.get('title', '')}")
    print(chunk.page_content[:300])
    print("---")
```

## 3. MarkdownHeaderTextSplitter (Section-Based Chunking)
Splits based on markdown headers preserved during scraping.
Best strategy for Wikipedia/Wikibooks — each chunk maps to one section,
and the header is stored as metadata for better RAG retrieval.

```python
headers_to_split_on = [
    ("#", "title"),
    ("##", "section"),
    ("###", "subsection"),
]

markdown_splitter = MarkdownHeaderTextSplitter(
    headers_to_split_on=headers_to_split_on
)

# Second-pass splitting keeps section structure but breaks long sections into
# retrieval-sized chunks that are better suited for embeddings and answer generation.
markdown_recursive_splitter = RecursiveCharacterTextSplitter(
    chunk_size=800,
    chunk_overlap=120,
    length_function=len,
    separators=["\n\n", "\n", ". ", " ", ""],
)

markdown_chunks = []
for doc in docs_for_chunking:
    text = doc.get('text', '')
    if not text:
        continue

    section_chunks = markdown_splitter.split_text(text)
    for section_chunk in section_chunks:
        base_metadata = dict(section_chunk.metadata)
        base_metadata['doc_title'] = doc.get('title', '')
        base_metadata['url'] = doc.get('url', '')

        child_chunks = markdown_recursive_splitter.create_documents(
            [section_chunk.page_content],
            metadatas=[base_metadata],
        )

        for subchunk_id, chunk in enumerate(child_chunks):
            chunk.metadata['subchunk_id'] = subchunk_id
            chunk.metadata['chunk_chars'] = len(chunk.page_content)
        markdown_chunks.extend(child_chunks)

print(f"Number of chunks: {len(markdown_chunks)}\n")
for i, chunk in enumerate(markdown_chunks[:10]):
    print(f"[Chunk {i}] {len(chunk.page_content)} chars | Metadata: {chunk.metadata}")
    print(chunk.page_content[:2000])
    print("---")
```

## 4. JSON Section-Based Chunking
Uses the existing `Summary` and `sections` fields directly instead of converting
the corpus into markdown headers first. This is the most natural fit for the
current Wikipedia-style JSON because section boundaries are already available in
the data. Long sections are further split with `RecursiveCharacterTextSplitter`
to keep chunk sizes retrieval-friendly.

```python
section_recursive_splitter = RecursiveCharacterTextSplitter(
    chunk_size=800,
    chunk_overlap=120,
    length_function=len,
    separators=["\n\n", "\n", ". ", " ", ""],
)

ignored_sections = {'see also', 'references', 'external links', 'further reading', 'notes'}
section_based_chunks = []

for doc in corpus:
    title = doc.get('title', '')
    url = doc.get('url', '')

    summary = (doc.get('Summary') or '').strip()
    if summary:
        summary_docs = section_recursive_splitter.create_documents(
            [summary],
            metadatas=[{
                'title': title,
                'doc_title': title,
                'url': url,
                'section': 'Summary',
            }],
        )
        for subchunk_id, chunk in enumerate(summary_docs):
            chunk.page_content = chunk.page_content.strip()
            chunk.metadata['subchunk_id'] = subchunk_id
            chunk.metadata['chunk_chars'] = len(chunk.page_content)
        section_based_chunks.extend(summary_docs)

    for sec in doc.get('sections', []) or []:
        sec_title = (sec.get('section_title') or '').strip()
        sec_text = (sec.get('section_text') or '').strip()
        if not sec_text:
            continue

        section_name = sec_title
        subsection_name = None
        if ' / ' in sec_title:
            section_name, subsection_name = [part.strip() for part in sec_title.split(' / ', 1)]

        if section_name.lower() in ignored_sections:
            continue

        metadata = {
            'title': title,
            'doc_title': title,
            'url': url,
            'section': section_name,
        }
        if subsection_name:
            metadata['subsection'] = subsection_name

        child_docs = section_recursive_splitter.create_documents(
            [sec_text],
            metadatas=[metadata],
        )
        for subchunk_id, chunk in enumerate(child_docs):
            chunk.page_content = chunk.page_content.strip()
            chunk.metadata['subchunk_id'] = subchunk_id
            chunk.metadata['chunk_chars'] = len(chunk.page_content)
        section_based_chunks.extend(child_docs)

print(f"Number of chunks: {len(section_based_chunks)}\n")
for i, chunk in enumerate(section_based_chunks[:10]):
    print(f"[Chunk {i}] {len(chunk.page_content)} chars | Metadata: {chunk.metadata}")
    print(chunk.page_content[:2000])
    print("---")
```

## 5. Semantic Chunking (Embedding-Based)
Uses sentence embeddings to detect topic shifts between adjacent text units.
This cell runs semantic chunking on the current corpus, which is set to the Cuisines80 blog JSON above.


```python
semantic_chunks = semantic_chunk_documents(
    texts=semantic_texts,
    metadatas=semantic_metadatas,
    model_name='BAAI/bge-small-en-v1.5',
    similarity_threshold=0.50,
    min_chunk_size=240,
    max_chunk_size=480,
    overlap_units=0,
)

print(f"Number of chunks: {len(semantic_chunks)}\n")
for i, chunk in enumerate(semantic_chunks[:5]):
    print(f"[Chunk {i}] {len(chunk.page_content)} chars | title={chunk.metadata.get('title', '')} | section={chunk.metadata.get('section_title', '')}")
    print(chunk.page_content[:300])
    print("---")
```

## 6. Comparison Summary
Comparing all five strategies on the same scraped Wikipedia text.

```python
char_sizes = [len(c.page_content) for c in char_chunks]
recursive_sizes = [len(c.page_content) for c in recursive_chunks]
markdown_sizes = [len(c.page_content) for c in markdown_chunks]
section_based_sizes = [len(c.page_content) for c in section_based_chunks]
semantic_sizes = [len(c.page_content) for c in semantic_chunks]

print(f"{'Strategy':<35} {'Num Chunks':<15} {'Avg Size':<15} {'Min':<10} {'Max'}")
print("-" * 85)
print(f"{'CharacterTextSplitter':<35} {len(char_chunks):<15} {sum(char_sizes)/len(char_sizes):<15.0f} {min(char_sizes):<10} {max(char_sizes)}")
print(f"{'RecursiveCharacterTextSplitter':<35} {len(recursive_chunks):<15} {sum(recursive_sizes)/len(recursive_sizes):<15.0f} {min(recursive_sizes):<10} {max(recursive_sizes)}")
print(f"{'MarkdownHeaderTextSplitter':<35} {len(markdown_chunks):<15} {sum(markdown_sizes)/len(markdown_sizes):<15.0f} {min(markdown_sizes):<10} {max(markdown_sizes)}")
print(f"{'JSON Section-Based':<35} {len(section_based_chunks):<15} {sum(section_based_sizes)/len(section_based_sizes):<15.0f} {min(section_based_sizes):<10} {max(section_based_sizes)}")
print(f"{'Semantic Chunking':<35} {len(semantic_chunks):<15} {sum(semantic_sizes)/len(semantic_sizes):<15.0f} {min(semantic_sizes):<10} {max(semantic_sizes)}")
```

## 7. Conclusion

Based on the scraped South Asian cuisine corpus:

- **CharacterTextSplitter**: Simple but may cut mid-sentence or mid-paragraph
- **RecursiveCharacterTextSplitter**: Smarter, respects natural text boundaries better
- **MarkdownHeaderTextSplitter + Recursive split**: Useful when the corpus is already stored as markdown or when you want a header-aware baseline
- **JSON section-based chunking**: Best fit for the current corpus because the JSON already stores section boundaries explicitly
- **Semantic chunking**: Useful when section boundaries are weak and you want embedding-based chunks that follow topic shifts more closely than fixed-size methods

**Decision**: Primary strategy → `JSON section-based chunking` + `RecursiveCharacterTextSplitter`  
**Fallback** (for unstructured text) → `Semantic chunking` or `RecursiveCharacterTextSplitter`
